### Duck Typing

> "–Ø–∫—â–æ –≤–æ–Ω–æ —Ö–æ–¥–∏—Ç—å —è–∫ –∫–∞—á–∫–∞ —ñ –∫—Ä—è–∫–∞—î —è–∫ –∫–∞—á–∫–∞ ‚Äî —Ü–µ –∫–∞—á–∫–∞."

Python –Ω–µ –ø–µ—Ä–µ–≤—ñ—Ä—è—î –¢–ò–ü –æ–±'—î–∫—Ç–∞ ‚Äî –≤—ñ–Ω –ø–µ—Ä–µ–≤—ñ—Ä—è—î –ü–û–í–ï–î–Ü–ù–ö–£:
```python
# –§—É–Ω–∫—Ü—ñ—è len() –ø—Ä–∞—Ü—é—î –∑ –±—É–¥—å-—á–∏–º, —â–æ –º–∞—î –º–µ—Ç–æ–¥ __len__
len("hello")     # str –º–∞—î __len__ ‚Üí 5
len([1, 2, 3])   # list –º–∞—î __len__ ‚Üí 3
len({"a": 1})    # dict –º–∞—î __len__ ‚Üí 1
```

–ù–µ–≤–∞–∂–ª–∏–≤–æ, —è–∫–∏–π —Ç–∏–ø ‚Äî –≤–∞–∂–ª–∏–≤–æ, —á–∏ –º–æ–∂–µ –æ–±'—î–∫—Ç "–∫—Ä—è–∫–∞—Ç–∏"!

````python
# Duck Typing –≤ –¥—ñ—ó

class Duck:
    def quack(self):
        return "–ö—Ä—è-–∫—Ä—è!"

class Person:
    def quack(self):
        return "–Ø —ñ–º—ñ—Ç—É—é –∫–∞—á–∫—É: –∫—Ä—è!"

class Robot:
    def quack(self):
        return "QUACK.WAV"

# –§—É–Ω–∫—Ü—ñ—è –Ω–µ –ø–µ—Ä–µ–≤—ñ—Ä—è—î —Ç–∏–ø ‚Äî —Ç—ñ–ª—å–∫–∏ –Ω–∞—è–≤–Ω—ñ—Å—Ç—å –º–µ—Ç–æ–¥—É quack()
def make_it_quack(thing):
    print(thing.quack())

# –í—Å—ñ —Ü—ñ –æ–±'—î–∫—Ç–∏ "–∫—Ä—è–∫–∞—é—Ç—å" ‚Äî –∑–Ω–∞—á–∏—Ç—å, –≤–æ–Ω–∏ –≤—Å—ñ "–∫–∞—á–∫–∏"!
make_it_quack(Duck())
make_it_quack(Person())
make_it_quack(Robot())

# –ê–ª–µ —è–∫—â–æ –æ–±'—î–∫—Ç –Ω–µ –≤–º—ñ—î –∫—Ä—è–∫–∞—Ç–∏...
class Cat:
    def meow(self):
        return "–ú—è—É!"

# make_it_quack(Cat())  # AttributeError: 'Cat' object has no attribute 'quack'
````

### –û–ø–µ—Ä–∞—Ç–æ—Ä–∏ (Operators)

````python
a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")    # –î–æ–¥–∞–≤–∞–Ω–Ω—è
print(f"a - b = {a - b}")    # –í—ñ–¥–Ω—ñ–º–∞–Ω–Ω—è
print(f"a * b = {a * b}")    # –ú–Ω–æ–∂–µ–Ω–Ω—è
print(f"a / b = {a / b}")    # –î—ñ–ª–µ–Ω–Ω—è (float)
print(f"a // b = {a // b}")  # –¶—ñ–ª–æ—á–∏—Å–µ–ª—å–Ω–µ –¥—ñ–ª–µ–Ω–Ω—è
print(f"a % b = {a % b}")    # –û—Å—Ç–∞—á–∞ –≤—ñ–¥ –¥—ñ–ª–µ–Ω–Ω—è
print(f"a ** b = {a ** b}")  # –ü—ñ–¥–Ω–µ—Å–µ–Ω–Ω—è –¥–æ —Å—Ç–µ–ø–µ–Ω—è
````

### F-strings (—Ñ–æ—Ä–º–∞—Ç–æ–≤–∞–Ω—ñ —Ä—è–¥–∫–∏)

F-strings ‚Äî –Ω–∞–π–∑—Ä—É—á–Ω—ñ—à–∏–π —Å–ø–æ—Å—ñ–± —Ñ–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è –≤ Python 3.6+:

````python
product = "–ö–∞–≤–∞"
price = 45.50
quantity = 3
total = price * quantity

# –ë–∞–∑–æ–≤–µ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è
print(f"–¢–æ–≤–∞—Ä: {product}")

# –§–æ—Ä–º–∞—Ç—É–≤–∞–Ω–Ω—è —á–∏—Å–µ–ª
print(f"–¶—ñ–Ω–∞: {price:.2f} –≥—Ä–Ω")
print(f"–í—Å—å–æ–≥–æ: {total:.2f} –≥—Ä–Ω")

# –í–∏—Ä–∞–∑–∏ –≤—Å–µ—Ä–µ–¥–∏–Ω—ñ f-string
print(f"3 √ó 45.50 = {3 * 45.50:.2f} –≥—Ä–Ω")

# –í–∏—Ä—ñ–≤–Ω—é–≤–∞–Ω–Ω—è
print(f"{'–¢–æ–≤–∞—Ä':<15}: {product}")
print(f"{'–¶—ñ–Ω–∞':<15}: {price:.2f} –≥—Ä–Ω")
````
